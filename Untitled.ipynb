## Data Cleaning

```python
#Check for missing values
df_mov.isnull().sum()
```

Missing values will be dropped and filled
Reasons
- Most missing values have are very few compared to the total amount of rows like imdb_id which has 10 missing rows, cast has 76 missing rows, overview has 4 and genres has 3. These will be dropped
- Columns like homepage, tagline and keywords will be filled with 'No Information'; the rows are too large to drop

```python
df_mov.fillna({'homepage':'No Information', 'tagline':'No Information', 'keywords':'No Information', 'production_companies':'No Information'}, inplace=True)
```

```python
#View the dataset to see if change was effective
df_mov.isnull().sum()
```

```python
# Dropping missing values
df_mov.dropna(inplace=True)
```

```python
#View the dataset to see if change was effective
df_mov.isnull().sum()
```

All missing values have been dropped 
Next, features that won't be needed will be dropped

```python
#View the dimensions
df_mov.shape
```

```python
#Drop columns that won't be needed
df_mov.drop(labels=['homepage', 'keywords'], inplace=True, axis=1)
```

```python
#view to see if change was effective
df_mov.columns
```

```python
#view the datatypes
df_mov.dtypes
```

```python
#view dimensions
df_mov.shape
```

```python
#change release date to datetime series
df_mov['release_date'] = pd.to_datetime(df_mov['release_date'])
```

```python
#view new data types
df_mov.dtypes
```

```python
df_mov['release_date'].dtypes
```

Check for duplicates

```python
df_mov.duplicated().sum()
```

```python
# Drop duplicates
df_mov.drop_duplicates(inplace=True)
```

```python
#confirm change
df_mov.duplicated().sum()
```

#### Next, Get top 5 movies produced in 2014
